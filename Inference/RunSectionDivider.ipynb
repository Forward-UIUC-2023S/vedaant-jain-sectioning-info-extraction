```python
from SectionDivider import SectionDivider
```

```python
import gensim.downloader as api  
w2v_model = api.load("word2vec-google-news-300")
```

```python
sec_divide = SectionDivider(w2v_model = w2v_model)
```

```python
from PIL import Image
import numpy as np
from numpy import asarray
image = Image.open('/content/gdrive/MyDrive/Resume_Info_Extraction/title_annotated/images/3-01_d024.jpg')
image = np.asarray(image)
```

```python
sections, texts = sec_divide.GetSections(image)
```

```python
texts
```

```python
import matplotlib.pyplot as plt
import math
fig, axes = plt.subplots(2, math.ceil(len(sections)/2), figsize=(19, 21))
axes = np.ravel(axes)
for i, section in enumerate(sections):
  axes[i].imshow(section)

plt.show()
```
